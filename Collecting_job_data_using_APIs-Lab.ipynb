```python
api_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json"

# Function to get number of jobs for a specific technology
def get_number_of_jobs_T(technology):
    """
    Returns the number of job postings for a given technology/skill
    """
    # Fetch data from API
    response = requests.get(api_url)
    data = response.json()
    
    # Count jobs containing the technology in Key Skills
    count = 0
    for job in data:
        if 'Key Skills' in job and job['Key Skills']:
            # Check if the technology appears in the key skills (case-insensitive)
            if technology.lower() in job['Key Skills'].lower():
                count += 1
    
    return technology, count
```

Find the number of jobs postings for each of the location in the above list.
Write the location name and the number of jobs postings into the excel spreadsheet.


```python
#your code goes here
# your code goes here
# API endpoint
api_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json"

# List of locations
locations = ["Los Angeles", "New York", "San Francisco", "Washington DC", "Seattle", "Austin", "Detroit"]

# Fetch data from API
response = requests.get(api_url)
data = response.json()

# Function to count jobs for a location
def get_job_count(location):
    count = 0
    for job in data:
        if 'Location' in job and job['Location']:
            if location.lower() in job['Location'].lower():
                count += 1
    return count

# Add headers
ws.append(["Location", "Number of Job Postings"])

# Format headers
for col in ['A', 'B']:
    ws[f'{col}1'].font = Font(bold=True, color="FFFFFF")
    ws[f'{col}1'].fill = PatternFill(start_color="366092", end_color="366092", fill_type="solid")
    ws[f'{col}1'].alignment = Alignment(horizontal="center")

# Add data for each location
row = 2
for location in locations:
    job_count = get_job_count(location)
    ws[f'A{row}'] = location
    ws[f'B{row}'] = job_count
    ws[f'B{row}'].number_format = '#,##0'
    row += 1

# Adjust column widths
ws.column_dimensions['A'].width = 20
ws.column_dimensions['B'].width = 22
```

#### In the similar way, you can try for below given key skills and results  can be stored in an excel sheet.


Collect the number of job postings for the following key skills using the API:

*   C
*   C#
*   C++
*   Java
*   JavaScript
*   Python
*   Scala
*   Oracle
*   SQL Server
*   MySQL Server
*   PostgreSQL
*   MongoDB


```python
# your code goes here
# your code goes here
import requests
import json
from openpyxl import Workbook
from openpyxl.styles import Font, PatternFill, Alignment

# API endpoint
api_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json"

# List of key skills to analyze
key_skills = [
    "C",
    "C#",
    "C++",
    "Java",
    "JavaScript",
    "Python",
    "Scala",
    "Oracle",
    "SQL Server",
    "MySQL Server",
    "PostgreSQL",
    "MongoDB"
]

# Fetch data from API
print("Fetching job data from API...")
response = requests.get(api_url)
data = response.json()
print(f"Total job records fetched: {len(data)}")

# Function to count jobs for a specific skill
def get_job_count_by_skill(skill):
    count = 0
    for job in data:
        if 'Key Skills' in job and job['Key Skills']:
            # Case-insensitive search for the skill
            if skill.lower() in job['Key Skills'].lower():
                count += 1
    return count

# Create workbook and worksheet
wb = Workbook()
ws = wb.active
ws.title = "Job Postings by Skill"

# Add headers
headers = ["Key Skill", "Number of Job Postings"]
ws.append(headers)

# Format headers
header_font = Font(bold=True, color="FFFFFF")
header_fill = PatternFill(start_color="366092", end_color="366092", fill_type="solid")
header_alignment = Alignment(horizontal="center")

for col in ['A', 'B']:
    ws[f'{col}1'].font = header_font
    ws[f'{col}1'].fill = header_fill
    ws[f'{col}1'].alignment = header_alignment

# Collect data for each skill
print("\n" + "="*60)
print("Job Postings by Key Skill:")
print("="*60)

row = 2
skill_counts = []

for skill in key_skills:
    job_count = get_job_count_by_skill(skill)
    skill_counts.append((skill, job_count))
    ws[f'A{row}'] = skill
    ws[f'B{row}'] = job_count
    ws[f'B{row}'].number_format = '#,##0'
    ws[f'B{row}'].alignment = Alignment(horizontal="right")
    print(f"{skill:<20} {job_count:>8,} job postings")
    row += 1

# Sort data by job count (highest to lowest) for better visualization
# Create a new sheet with sorted data
ws_sorted = wb.create_sheet("Sorted by Job Count")
ws_sorted.append(["Key Skill", "Number of Job Postings"])

# Format headers in sorted sheet
for col in ['A', 'B']:
    ws_sorted[f'{col}1'].font = header_font
    ws_sorted[f'{col}1'].fill = header_fill
    ws_sorted[f'{col}1'].alignment = header_alignment

# Sort skills by job count descending
sorted_skills = sorted(skill_counts, key=lambda x: x[1], reverse=True)

row = 2
for skill, count in sorted_skills:
    ws_sorted[f'A{row}'] = skill
    ws_sorted[f'B{row}'] = count
    ws_sorted[f'B{row}'].number_format = '#,##0'
    ws_sorted[f'B{row}'].alignment = Alignment(horizontal="right")
    row += 1

# Add total row
total_row = row
ws_sorted[f'A{total_row}'] = "TOTAL"
ws_sorted[f'B{total_row}'] = f"=SUM(B2:B{total_row-1})"
ws_sorted[f'A{total_row}'].font = Font(bold=True)
ws_sorted[f'B{total_row}'].font = Font(bold=True)
ws_sorted[f'B{total_row}'].number_format = '#,##0'

# Adjust column widths for both sheets
for sheet in [ws, ws_sorted]:
    sheet.column_dimensions['A'].width = 20
    sheet.column_dimensions['B'].width = 22

# Save to Excel file
output_filename = "skills-job-postings.xlsx"
wb.save(output_filename)

print("="*60)
print(f"\n✅ Successfully saved to '{output_filename}'")

# Display statistics
total_jobs = sum(count for _, count in skill_counts)
print(f"\n📊 Summary Statistics:")
print(f"Total job postings across all skills: {total_jobs:,}")
print(f"Average job postings per skill: {total_jobs/len(key_skills):,.0f}")
print(f"Skill with most job postings: {sorted_skills[0][0]} ({sorted_skills[0][1]:,})")
print(f"Skill with fewest job postings: {sorted_skills[-1][0]} ({sorted_skills[-1][1]:,})")

# Optional: Display top 5 skills
print(f"\n🏆 Top 5 Most In-Demand Skills:")
for i in range(min(5, len(sorted_skills))):
    print(f"   {i+1}. {sorted_skills[i][0]}: {sorted_skills[i][1]:,} job postings")
```
